Load users.csv into dataframe

Load ratings.csv into dataframe

```python
ratings = pd.read_csv("datasets/movielens_csv/ratings.csv")
ratings.head(5)
```

Load movies.csv

```python
movies  = pd.read_csv("datasets/movielens_csv/movies.csv")
movies.head(5)
```

```python
users.head(5)
ratings.head(5)
movies.head(5)
ratings
```

Join ratings with user using user_id as the join key

```python
pd.merge(ratings, users, on="user_id")
```

Join all three tables to get list of movies the users watched

```python
data = pd.merge(pd.merge(ratings, users), movies)
data
```

```python
data.iloc[0]
```

Get average watcher by gender for each movie

```python
mean_ratings = data.pivot_table("rating", index="title",
                                columns="gender", aggfunc="mean")
mean_ratings.head(100)
```

Group data by title and count the number of ratings

```python
ratings_by_title = data.groupby("title").size()
ratings_by_title.head()
```

Filter movies rating greater than 250

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles
```

Select movie title ratings by using active_titles index in the selection box

```python
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings
```

```python
mean_ratings = mean_ratings.rename(index={"Seven Samurai (The Magnificent Seven) (Shichinin no samurai) (1954)":
                           "Seven Samurai (Shichinin no samurai) (1954)"})
```

Most popular movies by females

```python
top_female_ratings = mean_ratings.sort_values("F", ascending=False)
top_female_ratings.head()
```

Get movies that are both popular among female, the lower the diff the more the movie is popular among women

```python
mean_ratings["diff"] = mean_ratings["M"] - mean_ratings["F"]
```

```python
sorted_by_diff = mean_ratings.sort_values("diff")
sorted_by_diff.head()
```

Reverse diff, the higher the diff the higher it popular among men

```python
sorted_by_diff[::-1].head()
```

Get diasgreement rating by using standard deviation of ratings

```python
rating_std_by_title = data.groupby("title")["rating"].std()
rating_std_by_title = rating_std_by_title.loc[active_titles]
rating_std_by_title.head()
```

```python
rating_std_by_title.sort_values(ascending=False)[:10]
```

Genres values are separated by pipe |, using split we treat | as a delimiter and split the values into a list.

```python
movies["genres"].head()
movies["genres"].head().str.split("|")
movies["genre"] = movies.pop("genres").str.split("|")
movies.head()
```

The popped value turn into a new row

```python
movies_exploded = movies.explode("genre")
movies_exploded[:10]
```

Merge exploded and popped up genre into rating and create a new csv file with that data

```python
ratings_with_genre = pd.merge(pd.merge(movies_exploded, ratings), users)
ratings_with_genre.to_csv("datasets/movielens_csv/ratings_with_genre.csv", index=False)
```

```python

genre_ratings = (ratings_with_genre.groupby(["genre", "age"])
                 ["rating"].mean()
                 .unstack("age"))
genre_ratings[:30]
```
